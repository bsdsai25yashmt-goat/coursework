```python
#Q1 PART - 1
def gcd(m, n):
    while n != 0:
        m, n = n, m % n
    return m
print(gcd(48, 18))
```

```python
#Q1 PART - 2

"""If negative numbers are provided, we usually convert them to absolute values so the function
always returns a positive GCD."""
# FOR EXAMPLE

print(gcd(-20, -8))
```

```python
#Q2 PART - 1
def euler_limit(n):
    return (1 + 1/n)**n
print(euler_limit(1000))
```

```python
#Q2 PART - 2
def euler_limit(n):
    return (1 + 1/n)**n
print(euler_limit(10^16))
```

```python
#Q2 PART - 3
def factorial(x):
    if x == 0:
        return 1
    else:
        return x * factorial(x-1)
        
def euler_sum(n):
    if n == 0:
        return 0.0
    total = 0.0
    
    for i in range(n):
        total += 1 / factorial(i)
        
    return total

print(euler_sum(0))
print(euler_sum(1))
print(euler_sum(2))
print(euler_sum(4))
```

```python
#Q2 PART - 4
import math

def euler_approx(epsilon):
    e = 0
    i = 0
    
    while True:
        term = 1 / math.factorial(i)
        if term < epsilon:
            break
        e += term
        i += 1
        
    return e

print(euler_approx(0.0001))
```

```python
#Q2 PART - 5
import math
def print_euler_sum_error(n):
    for k in range(1, n+1):
        approx = euler_sum(k)
        error = approx - math.exp(1)
        print(error)

def print_euler_lim_error(n):

    for k in range(1,n+1):
        approx = euler_limit(k)
        error = approx - math.exp(1)
        print(error)

print_euler_sum_error(5)
print()
print_euler_lim_error(5)
```

```python
#Q2 PART - 6

"""Between the two methods, the approximation using euler_sum() is better.
This is because euler_sum() uses the series expansion of e and each additional term
makes the result closer to the actual value. So even with a small number of terms, we
already get a good approximation.
On the other hand, the euler_limit method uses the expression ((1 + 1/n)^n). This
method converges more slowly and needs very large values of n to get close to the true
value of e. For very large n, floating-point precision can also affect the result.
So overall, euler_sum() gives a more accurate and reliable approximation of (e)**
compared to euler_limit.
```

```python
#Q3 PART - 1

def power_of_2(n):

    while n > 1:
        if n % 2 != 0:
            return False
        n = n // 2

    return True

print(power_of_2(5))
print(power_of_2(8))
```

```python
#Q3 PART - 2

def is_power(a, n):
    while n > 1:
        if n % a != 0:
            return False
        n = n // a
    return True

print(is_power(2, 8))
print(is_power(5, 9))        
```
